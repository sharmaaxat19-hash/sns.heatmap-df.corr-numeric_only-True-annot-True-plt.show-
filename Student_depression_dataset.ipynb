```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
df=pd.read_csv("n.csv")
```

```python
df.head()
```

```python
dept_count = df['Degree'].value_counts()

plt.figure(figsize=(8,5))
dept_count.plot(kind='bar')
plt.title("Number of Degree in Each prosession")
plt.xlabel("Profession")
plt.ylabel("Degree Count")
plt.show()
```

```python
dept_count = df['City'].value_counts()

plt.figure(figsize=(8,5))
dept_count.plot(kind='bar')
plt.title("City-wise Degree Count")
plt.xlabel("City")
plt.ylabel("Number of Degree")
plt.show()

print("City with Highest Degree:")
print(dept_count.idxmax())
```

```python
df=pd.read_csv("n.csv")
df['Academic Pressure'].value_counts().plot(kind='pie')
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Age'], bins=10)
plt.title("Age Distribution of Profession")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
gender_count = df['Gender'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(gender_count,
        labels=gender_count.index,
        autopct='%1.1f%%')

plt.title("Gender Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Profession'],
            df['Financial Stress'])

plt.title("Profession vs Financial Stress")
plt.xlabel("Profession")
plt.ylabel("Financial Stress")
plt.show()
```

```python
role_count = df['Job Satisfaction'].value_counts()

plt.figure(figsize=(10,5))
role_count.plot(kind='bar')
plt.title("Profession by Job Satisfaction")
plt.xlabel("Job Satisfaction")
plt.ylabel("Profession Count")
plt.xticks(rotation=45)
plt.show()
```

```python
top_cgpa = df.sort_values(by='CGPA', ascending=True).head(10)

plt.figure(figsize=(8,5))
plt.barh(top_cgpa['Profession'],
         top_cgpa['CGPA'])

plt.title("Top 10 Highest CGPA Profession")
plt.xlabel("CGPA")
plt.ylabel("Profession Name")
plt.show()
```

```python
top_cgpa = df.sort_values(
    by='CGPA',
    ascending=False
).head(10)

plt.figure(figsize=(10,5))
plt.bar(top_cgpa['Degree'],
        top_cgpa['CGPA'])

plt.title("Top CGPA Degree")
plt.xlabel("Degree")
plt.ylabel("CGPA")
plt.xticks(rotation=45)
plt.show()

```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Profession'],
            df['Financial Stress'])

plt.title("Profession vs Financial Stress")
plt.xlabel("Profession")
plt.ylabel("Financial Stress")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Age'], bins=10)
plt.title("Age Distribution of Profession")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Sleep Duration'],
         bins=5)

plt.title("Sleep Duration Distribution")
plt.xlabel("Sleep Duration")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```
